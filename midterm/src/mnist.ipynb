```python
import idx2numpy

def load_MINST_dataset(path):
    train_data = idx2numpy.convert_from_file(f"{path}/train-images-idx3-ubyte")
    train_label = idx2numpy.convert_from_file(f"{path}/train-labels-idx1-ubyte")
    test_data = idx2numpy.convert_from_file(f"{path}/t10k-images-idx3-ubyte")
    test_label = idx2numpy.convert_from_file(f"{path}/t10k-labels-idx1-ubyte")
    return train_data, train_label, test_data, test_label

def flatten_images(train_data, test_data):
    return train_data.reshape(train_data.shape[0], -1), test_data.reshape(test_data.shape[0], -1)
```

```python
import numpy as np

path = "../data/MNIST"
X_train, y_train, X_test, y_test = load_MINST_dataset(path)
print("Before flattening")
print(X_train.shape)
print(y_train.shape)

X_train, X_test = flatten_images(X_train, X_test)
print("After flattening")
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardization
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

pca_50 = PCA(n_components=50)
pca_100 = PCA(n_components=100)
pca_200 = PCA(n_components=200)

X_train_50 = pca_50.fit_transform(X_train)
X_test_50 = pca_50.transform(X_test)

X_train_100 = pca_100.fit_transform(X_train)
X_test_100 = pca_100.transform(X_test)

X_train_200 = pca_200.fit_transform(X_train)
X_test_200 = pca_200.transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

linear_svc = SVC(kernel='linear', C=1.0)
rbf_svc = SVC(kernel='rbf', gamma = 'scale')
poly_svc = SVC(kernel='poly', degree = 3, gamma= 'scale')

# From here figure out and fine tune all the different models manually and PCA versions using training error, test error, and other techniques
linear_svc.fit(X_train_50, y_train)

y_pred = linear_svc.predict(X_test_50)
accuracy = (y_test == y_pred).mean()
cm = confusion_matrix(y_test, y_pred)
```
